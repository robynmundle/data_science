### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.
- first table is in USD
- second table actually consists of 3 smaller tables. First 9 rows are for USD, next 8 rows are for GBP and final 7 are for EUR.

```python
col = sales2.Title.str.extract(r'\((.*)\)')
sales2['currency'] = col.fillna(method='bfill')
sales2
```

```python
together = pd.merge(sales2, sales1, left_on='Title', right_on='Book title', how='outer')
together
```

```python
together['total'] = (together.fillna(0)['Units sold']*together.fillna(0)['List price']) + (together.fillna(0)['Number sold']*together.fillna(0)['Sales price'])
together
```

```python
together.groupby(['Title','currency','total']).size()
```
