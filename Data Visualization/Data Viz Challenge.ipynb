```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import numpy as np
import calendar


df = pd.read_csv('data/results.csv', sep=',')
df.head()
```

## Task
Which teams scored the largest number of goals in FIFA World Cup?

```python
fifa = df['tournament'] == 'FIFA World Cup'
home_goals = df[fifa][['home_team','home_score']].groupby('home_team').sum()
home_goals.reset_index(inplace=True)
away_goals = df[fifa][['away_team','away_score']].groupby('away_team').sum()
away_goals.reset_index(inplace=True)

goals = home_goals.merge(away_goals, how='outer', left_on='home_team', right_on='away_team')
```

```python
goals['goals'] = goals.home_score + goals.away_score
top_15 = goals.sort_values(by='goals', ascending=False).head(15)
```

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.set_style("darkgrid")
ax = sns.barplot(x=top_15["home_team"], y=top_15['goals'])
# Change the default axis labels
ax.set_title('Top 15 Teams Scoring Goals in FIFA World Cup')
ax.set_xlabel('Team')
ax.set_ylabel('Total Goals Scored')
plt.show()
```

## Task
What is the number of matches played in each tournament throughout history?

```python
df['year'] = pd.DatetimeIndex(df['date']).year    

df.head()
```

```python
matches = pd.DataFrame(df.groupby(['tournament','year']).agg('size'))
matches.reset_index(inplace=True)
matches = matches.rename(columns={0:'nb_games'})
matches.sort_values(by='nb_games', ascending=False).head()
```

```python
tournaments = matches.groupby('tournament').sum()
tournaments = tournaments.sort_values('nb_games',ascending=False).head(10)
tournaments.reset_index(inplace=True)
tournaments
```

```python
# HORIZONTAL BAR PLOT
sns.set_theme(style="whitegrid")



# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot the total games
sns.set_color_codes("pastel")
sns.barplot(x="nb_games", y="tournament", 
            data=tournaments, color="b")

# Add a legend and informative axis label
ax.set(xlim=(0, 20000), ylabel="",
      xlabel="Number of Matches Played",
      title='Matches Played of Top 10 Tournaments')
sns.despine(left=True, bottom=True)
```

## Task
Which teams are the most successful ones? (winning percentage)

```python
conditions = [
    (df['home_score'] > df['away_score']), (df['home_score'] < df['away_score'])
]
choices = [
    True,
    False
]
df['homewon'] = np.select(conditions,choices,default=np.nan)
conditions = [
    (df['home_score'] < df['away_score']), (df['home_score'] > df['away_score'])
]
choices = [
    True,
    False
]
df['awaywon'] = np.select(conditions,choices,default=np.nan)
```

```python
home_count = pd.DataFrame(df.groupby('home_team').agg('size'))
home_count.reset_index(inplace=True)
home_count = home_count.rename(columns={'home_team': 'team_name', 0:'nb_homegames'})

home_wins = df[['home_team','homewon']].groupby('home_team').sum()
home_wins.reset_index(inplace=True)
home_wins = home_wins.rename(columns={'home_team': 'team_name'})

home = home_count.merge(home_wins, on='team_name')


away_count = pd.DataFrame(df.groupby('away_team').agg('size'))
away_count.reset_index(inplace=True)
away_count = away_count.rename(columns={'away_team': 'team_name', 0:'nb_awaygames'})

away_wins = df[['away_team','awaywon']].groupby('away_team').sum()
away_wins.reset_index(inplace=True)
away_wins = away_wins.rename(columns={'away_team': 'team_name'})

away = away_count.merge(away_wins, on='team_name')


teams = home.merge(away, on='team_name')
teams['total_games'] = teams.nb_homegames + teams.nb_awaygames
teams['games_won'] = teams.homewon + teams.awaywon
teams['games_lost'] = teams.total_games - teams.games_won
teams['win_rate'] = teams.games_won / teams.total_games
teams['lose_rate'] = teams.games_lost / teams.total_games
games = teams['total_games'] > 100
teams = teams[games].sort_values(by='win_rate', ascending=False).head(20)
teams
```

```python
# HORIZONTAL BAR PLOT
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot the total games
sns.set_color_codes("pastel")
sns.barplot(x="win_rate", y="team_name", 
            data=teams, color="g")

# Add a legend and informative axis label
ax.set(xlim=(0, 1), ylabel="",
      xlabel="Win Rate",
      title='Successful Football Teams')
sns.despine(left=True, bottom=True)
```

## Task
Which teams played against each other the most?

```python
df.head()
```

```python
from collections import Counter
```

```python
# for loop in each row by calling the shape, capture each team occurence
two_teams = []
for i in range(df.shape[0]):
    teams = [df.loc[i,'home_team'], df.loc[i,'away_team']]
    teams.sort()
    two_teams.append(tuple(teams))
```

```python
count_teams = Counter(two_teams).most_common()
teamnames = [x[0] for x in count_teams]
gamesplayed = [x[1] for x in count_teams]
matchup = [x[0] + " x " + x[1] for x in teamnames]
```

```python
# HORIZONTAL BAR PLOT
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot the total games
sns.set_color_codes("dark")
sns.barplot(x=gamesplayed[:10], y=matchup[:10],
           color='r')

# Add a legend and informative axis label
ax.set(title='Top 10 Most Common Team Match-Ups')
sns.despine(left=True, bottom=True)
```

```python
h_a = df[['home_team','away_team','date']].groupby(['home_team','away_team']).count()
h_a.reset_index(inplace=True)
h_a = h_a.rename(columns={'home_team':'team_1', 'away_team':'team_2','date':'h_a_count'})

a_h = df[['away_team','home_team','date']].groupby(['away_team','home_team']).count()
a_h.reset_index(inplace=True)
a_h = a_h.rename(columns={'home_team':'team_2', 'away_team':'team_1','date':'a_h_count'})

matches = h_a.merge(a_h, on=['team_1','team_2'])
matches
```

```python
matches['games'] = matches.h_a_count + matches.a_h_count
matches['Team Match-Up'] = matches.team_1.astype(str) + ' x ' + matches.team_2.astype(str)
matches = matches[['Team Match-Up','games']].sort_values('games',ascending=False).head(20)
matches
```
