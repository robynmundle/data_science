```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

* Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>
* hyperbolic paraboloid equation : z = x*x - y*y

```python
# wireframe

#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))
# create an empty canvas
ax = plt.axes(projection='3d')

# divide the interval <-10,10> into 500 parts
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

Z = (X*X) - (Y*Y)


# a wireframe
ax.plot_wireframe(X, Y, Z, color='r')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))
# create an empty canvas
ax = plt.axes(projection='3d')

# divide the interval <-10,10> into 500 parts
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

z = (x*x) + (y*y)

# a surface plot
ax.plot_surface(X, Y, Z, cmap='jet')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* Create also contour projection on the plane z=0 (HINT: use parameter 'zdir' of contour plot)
* eliptical paraboloid : z = x*x + y*y

```python
# contour plot
fig, ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

ax.contour(X, Y, Z, 40, zdir='z', cmap='jet')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_zlim(-100, 200)
plt.show()
```
