```python
# import plotly graph objects
import plotly.graph_objs as go

# import pandas
import pandas as pd

# import numpy
import numpy as np
```

# Objective: Create a scatterplot of 1000 random data points. 
x-axis values should come from a normal distribution using np.random.randn(1000).  
y-axis values should come from a uniform distribution over [0,1) using np.random.rand(1000)

```python
# obtain x and y values:
x_values = np.random.randn(1000)    # 1000 sample of norm dist
y_values = np.random.rand(1000) # 1000 values from the interval <0,1>, equidistantly divided

# define a data variable
data = go.Scatter(
    x = x_values,
    y = y_values,
     mode = 'markers'
)

# define the layout, and include a title and axis labels
layout = go.Layout(
    title = 'Scatterplot of 1000 random data points',
    xaxis_title="Normal Distribution",
    yaxis_title="Uniform Distribution",
)

# Create a fig from data and layout, and plot the fig
fig = go.Figure(data=data,layout=layout)
fig.show()
```

# Objective: Create a stacked bar chart. Note that questions appear in the index (and should be used for the x-axis), while responses appear as column labels.  
Extra Credit: make a horizontal bar chart! See https://plot.ly/python/horizontal-bar-charts/ for extra credit help.

```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/mocksurvey.csv',index_col=0)
df.head()
```

```python
# create traces using a list comprehension:
# trace1 - Q1
trace0 = go.Bar(
    x=df.index,
    y=df['Strongly Agree'],
    name = 'Strongly Agree'
)
trace1 = go.Bar(
    x=df.index,
    y=df['Somewhat Agree'],
    name = 'Somewhat Agree'
)
trace2 = go.Bar(
    x=df.index,
    y=df['Neutral'],
    name = 'Neutral'
)
trace3 = go.Bar(
    x=df.index,
    y=df['Somewhat Disagree'],
    name = 'Somewhat Disagree'
)
trace4 = go.Bar(
    x=df.index,
    y=df['Strongly Disagree'],
    name = 'Strongly Disagree'
)
data = [trace0, trace1, trace2, trace3, trace4]
# create a layout, remember to set the barmode here

layout = go.Layout(
    title='Standard Responses per Question',
    barmode = 'stack'
)

# create a fig from data & layout, and plot the fig.
fig = go.Figure(data=data, layout=layout)
fig.show()

```

```python
fig = go.Figure()
# create traces using a list comprehension:
# trace1 - Q1
fig.add_trace(go.Bar(
    x=df['Strongly Agree'],
    y=df.index,
    name = 'Strongly Agree',
    orientation='h'
))
fig.add_trace(go.Bar(
    x=df['Somewhat Agree'],
    y=df.index,
    name = 'Somewhat Agree',
    orientation='h'
))
fig.add_trace(go.Bar(
    x=df['Neutral'],
    y=df.index,
    name = 'Neutral',
    orientation='h'
))
fig.add_trace(go.Bar(
    x=df['Somewhat Disagree'],
    y=df.index,
    name = 'Somewhat Disagree',
    orientation='h'
))
fig.add_trace(go.Bar(
    x=df['Strongly Disagree'],
    y=df.index,
    name = 'Strongly Disagree',
    orientation='h'
))
# create a layout, remember to set the barmode here

fig.update_layout(
    title='Standard Responses per Question',
    barmode = 'stack'
)

# create a fig from data & layout, and plot the fig.
fig.show()

```

# Objective: Create a histogram that plots the 'length' field from dataset. Set the range from 0 to 1, with a bin size of 0.02

```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/abalone.csv')
df.head()
```

```python
# create a data variable:
data = [
    go.Histogram(
        # length
        x = df['length'],
        name = 'Length',
        xbins=dict( # bins used for histogram
            start=0,
            end=1,
            size=0.02
        )
    )]

# add a layout
layout = go.Layout(
    title="Lengths Distribution",
    xaxis_title="Lengths",
    yaxis_title="Count",
)

# create a fig from data & layout, and plot the fig
fig = go.Figure(data=data, layout=layout)
fig.show()
```
